```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.football.org.il/team-details/team-games/?team_id=1005&season_id=20'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
games = soup.findAll('a', 'table_row link_url')
translateDict= {
    'תאריך': 'Date',
    'משחק': 'Game',
    'אצטדיון': 'Venue',
    'שעה': 'Time',
    'תוצאה': 'Result'
}

gamesParsed = []
for game in games:
    gameInfo = game.findAll('div')
    gameParsed = {}
    for info in gameInfo:
        gameParsed[translateDict[info.span.contents[0]]] = info.contents[1]
    gamesParsed.append(gameParsed)
    
gamesPd = pd.DataFrame(gamesParsed)
gamesPd.head()
```

```python
def PointsPerResult(lhs, rhs):
    if lhs>rhs:
        return 3, 0
    elif lhs==rhs:
        return 1, 1
    return 0, 3

def CalcPointsEarned(game, result, selectedTeam):
    splittedGame = game.split('-')
    splittedResult = result.split('-')
    homePoints, awayPoints = PointsPerResult(splittedResult[0], splittedResult[1])
    if selectedTeam==splittedGame[1].strip():
        return homePoints
    return awayPoints

def GoalsFor(game, result, selectedTeam):
    splittedGame = game.split('-')
    splittedResult = result.split('-')
    if selectedTeam==splittedGame[1].strip():
        return int(splittedResult[0])
    return int(splittedResult[1])

def GoalsAgainst(game, result, selectedTeam):
    splittedGame = game.split('-')
    splittedResult = result.split('-')
    if selectedTeam==splittedGame[1].strip():
        return int(splittedResult[1])
    return int(splittedResult[0])


gamesPd['PointsEarned'] = gamesPd.apply(lambda row: CalcPointsEarned(row['Game'], row['Result'], 'מכבי חיפה'), axis=1)
gamesPd['GoalsFor'] = gamesPd.apply(lambda row: GoalsFor(row['Game'], row['Result'], 'מכבי חיפה'), axis=1)
gamesPd['GoalsAgainst'] = gamesPd.apply(lambda row: GoalsAgainst(row['Game'], row['Result'], 'מכבי חיפה'), axis=1)

gamesPd['PointsEarnedSum'] = gamesPd['PointsEarned'].cumsum()
gamesPd['GoalsForSum'] = gamesPd['GoalsFor'].cumsum()
gamesPd['GoalsAgainstSum'] = gamesPd['GoalsAgainst'].cumsum()
gamesPd
```
